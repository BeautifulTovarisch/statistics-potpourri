# Random Variables

A random variable $X$ in a sample space $\Omega$ takes an outcome as an argument and maps it to some (commonly numerical) value.

For example, let $\Omega$ consist of 6 names, and let $X : \Omega \to \mathbb{Z}$, be a random variable which counts the number of letters in each name (e.g $X(\text{Pete}) = 4$).

We would like to find the _Probability Distribution_ of $X$ given by:

\begin{equation}
  p(x) := P(X = x), \; \text{for} \; x \in \mathbb{Z}
\end{equation}

This reads:

> $p(x)$ is the probability distribution of random variable $X$.

In this example $p(x)$ is a _Probability Mass Function_ representing the probability distribution of discrete random variable $X$.

```julia
import Pkg

Pkg.add("Plots");
Pkg.add("PyPlot");
Pkg.add("StatsBase");
```

```julia
"""
Counting Names

This program models the random variable described above. A Monte Carlo simulation
chooses randomly from a list of names and computes the probability distribution
of given name lengths.
"""

using StatsBase, Plots; pyplot()

names = ["Mary", "Mel", "David", "John", "Kayley", "Anderson"]

X = 3:8
N = 10^6

distribution = counts([length(rand(names)) for _ in 1:N])/N

bar(X, distribution,
  ylims=(0, 1),
  xlabel="Name Length",
  ylabel="p(x)", legend=:none)
```
